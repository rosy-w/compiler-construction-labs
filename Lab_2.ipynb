###Building a Simple Lexical Analyzer using Flex and Bison

Flex is designed to transform an input file  usually in the format *.l* file. to a C program contained in a source file called *lex.yy.c*. This file contains C code that implements an FA that encodes all the rules and actions specified. The file provides an external function *yylex()* that will scan one token.

The flex.l file is as shown below:

Screenshot from 2022-09-25 20-11-14.png

In this case, the input file is *flex.l*. The flex command is used to read that file. This generates a *lex.yy.c* file. The *lex.yy.c* file is then compiled by the C compiler into a file called *a.out* by default. This file can be named by adding the '-o' tag to the command followed by the chosen name.This output from the compiler is a working lexical analyzer that will take an input stream of characters and produce a series of tokens.
