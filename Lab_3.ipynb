###Structure of the Flex input file
The flex input file is divided into these main sections:
1.   **Declarations.** This is an optional section.It is the topmost in the file and is taken from the normal C code.
2.   **Definitions**. This is where options are set for the scanner.Definitions to give names for regular expressions are set in this section. This aids in having simpler entries in the rules section.
3.   **Rules**. Here rules are created as specified patterns that should identify tokens and perform an action upon identifying the tokens. 
4.   **User Code**. User code that contains the code to run on the terminal and facilitate the collection of input.This section contains the main function and helper functions of the file.
