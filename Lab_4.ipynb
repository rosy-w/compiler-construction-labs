##Predictive parsing
**Definition**

Predictive parsing is a form of recursive-descent parsing where there is no backtracking. A predictive parser is a program that has a production for every non-terminal.

##LL1 Parsing
**Definition**

LL(1) Parsing is a method of top-down parsing where the next production is determined by looking ahead one terminal. 

**First() and Follow()**

The First() and Follow() are functions used to construct LL(1) parsers, among other top-down and bottom-up parsers.These functions produce the first and follow sets, which are are used to generate a predictive parsing table. For LL(1) grammars, the parsing tables entries uniquely identify a production.
